```python
class Process:
  def __init__(self, processId=0, burstTimes=0, burstIndex=0, currentState='ready',arrivalTime=0, totalCPUtime=0, rt=0, wt=0, tt=0, queue=1):
    self.processId = processId
    self.burstTimes = burstTimes
    self.burstIndex = burstIndex
    self.currentState = currentState
    self.arrivalTime = arrivalTime
    self.totalCPUtime = totalCPUtime
    self.rt = rt
    self.wt = wt
    self.tt = tt
    self.queue = queue

cpuTime = 0
idleTime = 0

# PROCESS INITIALIZATION
process = []
process.append(Process(1, [6, 21, 9, 28, 5, 26, 4, 22, 3, 41, 6, 45, 4, 27, 8 , 27, 3]))
process.append(Process(2, [19, 48, 16, 32, 17, 29, 6, 44, 8, 34, 21, 34, 19, 39, 10, 31, 7]))
process.append(Process(3, [12, 14, 6, 21, 3, 29, 7, 45, 8, 54, 11, 44, 9]))
process.append(Process(4, [11, 45, 5, 41, 6, 45, 8, 51, 4, 61, 13, 54, 11, 61, 10]))
process.append(Process(5, [16, 22, 15, 21, 12, 31, 14, 26, 13, 31, 16, 18, 12, 21, 10, 33, 11]))
process.append(Process(6, [20, 31, 22, 30, 25, 29, 11, 44, 17, 34, 18, 31, 6, 22, 16]))
process.append(Process(7, [3, 44, 7, 24, 6, 34, 5, 54, 4, 24, 7, 44, 6, 54, 5, 21, 6, 43, 4]))
process.append(Process(8, [15, 50, 4, 23, 11, 31, 4, 31, 3, 47, 5, 21, 8, 31, 6, 44, 9]))

finalized = []  # array for the finalized processes

def q1(ind):
  count = 0
  while True:
    process[ind].burstTimes[process[ind].burstIndex] -= 1
    global cpuTime
    cpuTime +=1
    count += 1
    if process[ind].burstTimes[process[ind].burstIndex] == 0:
      
      count = 0
      process[ind].arrivalTime = cpuTime + process[ind].burstTimes[process[ind].burstIndex + 1] # save new arrival time after I/O burst
      process[ind].totalCPUtime += process[ind].burstTimes[process[ind].burstIndex] # save new total CPU time for process
      process[ind].burstIndex += 2
      process.sort(key=lambda x: x.arrivalTime)

       # IMPORTANT: Sort processes by arrivalTime
      break
    if count == 6:
    
      process[ind].queue = 2
      process[ind].arrivalTime = cpuTime
      process[ind].totalCPUtime += cpuTime
      process.sort(key=lambda x: x.arrivalTime) # IMPORTANT: Sort processes by arrivalTime
      process.sort(key=lambda x: x.queue)
      break

def q2(ind):

  count = 0
  while True:
    process[ind].burstTimes[process[ind].burstIndex] -= 1
    global cpuTime
    cpuTime +=1
    count += 1
    if process[ind].burstTimes[process[ind].burstIndex] == 0:
      
      count = 0
      process[ind].arrivalTime = cpuTime + process[ind].burstTimes[process[ind].burstIndex + 1] # save new arrival time after I/O burst
      process[ind].totalCPUtime += process[ind].burstTimes[process[ind].burstIndex] # save new total CPU time for process
      process[ind].burstIndex += 2
      process.sort(key=lambda x: x.arrivalTime)
      process.sort(key=lambda x: x.queue)
       # IMPORTANT: Sort processes by arrivalTime
      break
    if count == 12:
    
      process[ind].queue = 3
      process[ind].arrivalTime = cpuTime
      process[ind].totalCPUtime += cpuTime
      process.sort(key=lambda x: x.arrivalTime) # IMPORTANT: Sort processes by arrivalTime
      process.sort(key=lambda x: x.queue)
def q3(ind):

  while True:
    global cpuTime
    for i in range(len(process)):
      if process[i].queue == 1:
        if cpuTime >= process[i].arrivalTime:
          process[ind].arrivalTime = cpuTime
          q1(i)
          return
    for i in range(len(process)):
      if process[i].queue == 2:
        if cpuTime >= process[i].arrivalTime:
          process[ind].arrivalTime = cpuTime
          q2(i)
          return
    
    process[ind].burstTimes[process[ind].burstIndex] -= 1
    cpuTime +=1

    if process[ind].burstTimes[process[ind].burstIndex] <= 0:
      process[ind].burstIndex += 1
      process[ind].arrivalTime = cpuTime + process[ind].burstTimes[process[ind].burstIndex] # save new arrival time after I/O burst
      process[ind].totalCPUtime += process[ind].burstTimes[process[ind].burstIndex] # save new total CPU time for process
      process[ind].burstIndex += 1
      process.sort(key=lambda x: x.arrivalTime)
      process.sort(key=lambda x: x.queue) # IMPORTANT: Sort processes by arrivalTime
      break

def printTimes():
  totalRT = 0
  totalWT = 0
  totalTT = 0
  utility = ((cpuTime-idleTime)/cpuTime)*100

  print("CPU Utilization: {}%\n".format(round(utility,2)))
  print("\tRT\tWT\tTT")
  for i in range(len(finalized)):
    finalized[i].wt = finalized[i].tt - finalized[i].totalCPUtime
    print("P{}:\t{}\t{}\t{}".format(finalized[i].processId, finalized[i].rt, finalized[i].wt, finalized[i].tt))
    totalRT += finalized[i].rt
    totalWT += finalized[i].wt
    totalTT += finalized[i].tt
  totalRT = totalRT/len(finalized)
  totalWT = totalWT/len(finalized)
  totalTT = totalTT/len(finalized)
  print("AVG:\t{}\t{}\t{}".format(totalRT, totalWT, totalTT))

def printProcesses():
  # CURRENT PROCESS
  print("Current Time:\t{}".format(cpuTime))
  print("Now running:\tP{}".format(process[0].processId))

  # READY QUEUE
  print("-----------------------------------")
  print("Ready Queue:\tProcess\tBurst\tQueue")
  for i in range(len(process)-1):
    if cpuTime >= process[i+1].arrivalTime:
      if process[i+1].burstIndex == 0:
        print("\t\tP{}\t{}\t{}".format(process[i+1].processId, process[i+1].burstTimes[process[i+1].burstIndex], process[i+1].queue))
      else:
        print("\t\tP{}\t{}\t{}".format(process[i+1].processId, process[i+1].burstTimes[process[i+1].burstIndex+1], process[i+1].queue))

  # if cpu time is less than arrival time then it is in I/O
  # I/O QUEUE
  print("-----------------------------------")
  print("In I/O:\t\tProcess\tI/O Burst")
  for i in range(len(process)):
    if cpuTime < process[i].arrivalTime:
      ioBurst = process[i].burstTimes[process[i].burstIndex]
      print("\t\tP{}\t{}".format(process[i].processId, ioBurst))    

  # if any finalized, then print completed processes
  # COMPLETED PROCESSES
  print("-----------------------------------")
  print("Completed:\tProcess")
  if finalized: 
    for i in range(len(finalized)):
      print("\t\tP{}".format(finalized[i].processId))
  else:
    print("\t\tNONE")
  print("***********************************\n")

while True:
  finalizedCheck = 0
  for i in range(len(process)):
    if process[i].currentState == 'finalized':
      finalizedCheck+=1
  #if all processes are finalized
  if finalizedCheck == len(process):
    print("Final CPU time: {}\n".format(cpuTime))
    finalized.sort(key=lambda x: x.processId)
    printTimes()
    break

  print("***********************************")
  
  try:
    if cpuTime < process[0].arrivalTime:
      idleTime += (process[0].arrivalTime - cpuTime)
      cpuTime = process[0].arrivalTime
    if process[0].burstIndex == 0:
      process[0].rt = cpuTime

    printProcesses()

    #QUEUE 1
    if process[0].queue == 1:
      q1(0)
      continue
    #QUEUE 2
    if process[0].queue == 2:
      q2(0)
      continue
    #QUEUE 3
    if process[0].queue == 3:
      q3(0)
      continue

    #process.sort(key=lambda x: x.arrivalTime) # IMPORTANT: Sort processes by arrivalTime
  except IndexError:
    process[0].currentState = 'finalized' # set the currentState to 'finalized'
    process[0].tt = cpuTime # store finalized cpu time in 'tt'
    temp = process[0] # store process in temp
    process.pop(0) # remove process from array
    finalized.append(temp) # add process to finalized array
```
